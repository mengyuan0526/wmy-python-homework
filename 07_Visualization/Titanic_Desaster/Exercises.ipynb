### Step 1. Import the necessary libraries

### Step 3. Assign it to a variable titanic 

```python
titanic=pd.read_csv("../../../../pandas_exercises_with_data/data/titanic.csv")
titanic
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId')
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
sex_m=(titanic['Sex']=="male").sum()
sex_f=(titanic['Sex']=="female").sum()
sex_p=[sex_m,sex_f]
plt.pie(sex_p,labels=["male","female"],colors=["green","blue"],startangle = 90,shadow=False)
plt.title("male/female proportion pie")
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.lmplot(x="Age",y="Fare",data=titanic,hue="Sex")
```

### Step 7. How many people survived?

### Step 8. Create a histogram with the Fare payed

```python
sns.distplot(titanic['Fare'],color='blue')
plt.xlim(0,300)
plt.ylim(0,0.04)
```
